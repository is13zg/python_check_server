## Задача 1

Значение выражения 7^103 + 20∙7^104 – 3∙7^57 + 97 записали в системе счисления с основанием 7. Сколько цифр 6 содержится в этой записи?

## Решение

```python
x = 7**103+20*7**104-3*7**57+97
a = []
while x>0:
  a = [x%7] + a
  x = x // 7
print(a.count(6))
```

## Задача 6

При каком наименьшем натуральном значении переменной x в выражении
81^20 – 9^x + 50 сумма цифр в девятеричной записи числа равна 138?

## Решение

```python

for n in range(1,100):
  x = 81**20 - 9**n + 50
  a = []
  while x>0:
    a = [x%9] + a
    x = x // 9
  if sum(a)==138:
    print(n, sum(a))

```

## Задача 7

Значение выражения 64^12 – 8^14 + x записали в восьмеричной системе счисления, при этом в записи оказалось 12 цифр 7 и одна единица. При каком наименьшем натуральном  x это возможно?

## Решение

```python

for n in range(1,1000):
  x = 64**12 - 8**14 + n
  a = []
  while x>0:
    a = [x%8] + a
    x = x // 8
  if a.count(7)==12 and a.count(1)==1:
    print(n)

```

## Задача 12

В некоторой системе счисления записи десятичных чисел 56 и 45 заканчиваются на 1. Определите основание системы счисления.

## Решение

```python
for n in range(2, 30):
    x = 56
    a = []
    while x > 0:
        a = [x % n] + a
        x = x // n

    y = 45
    b = []
    while y > 0:
        b = [y % n] + b
        y = y // n

    if a[-1] == 1 and b[-1] == 1:
        print(n)
```
